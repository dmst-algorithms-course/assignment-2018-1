# Η Πλάνη του Φάνη

Το «Βραβείο στις Οικονομικές Επιστήμες της Τράπεζας της Σουηδίας στη Μνήμη του Άλφρεντ Νόμπελ», κοινώς γνωστό και ως βραβείο Νόμπελ στα οικονομικά, δώθηκε το 2017 στον [Richard Thaler](https://en.wikipedia.org/wiki/Richard_Thaler), του πανεπιστημίου του Σικάγο, για τις συνεισφορές του στα [Συμπεριφορικά Οικονομικά (Behavioural Economics)](https://en.wikipedia.org/wiki/Behavioral_economics). Σύμφωνα λοιπόν με τη Wikipedia, τα Συμπεριφορικά Οικονομικά μελετούν τον αντίκτυπο των ψυχολογικών, κοινωνικών, γνωσιακών, και συναισθηματικών παραγόντων στις οικονομικές αποφάσεις του ατόμου και των θεσμών, όπως και τα αποτελέσματα των παραγόντων αυτών στον καθορισμό των τιμών, επενδυτικές αποδόσεις, καταμερισμό πόρων, κ.λπ. 

Εμείς θα δούμε πώς μπορούμε να διερευνήσουμε μια εφαρμογή Συμπεριφορικών Οικονομικών μέσω γράφων. Συγκεκριμένα, θα δούμε πώς ένα άτομο μπορεί να κάνει λάθος επιλογές κάτω από συγκεκριμένες συνθήκες. Ο ψυχολογικός παράγοντας που θα εξετάσουμε είναι η λεγόμενη *μεροληψία του παρόντος* (present bias). Η μεροληψία του παρόντος είναι η τάση που έχουμε να θεωρούμε τα κόστη και τα ωφέλη που έχουμε στο παρόν πιο σημαντικά από αυτά που θα έρθουν στο μέλλον. Για να δούμε δύο παραδείγματα:

* Πολλές φορές ένα χρηματικό κέρδος σήμερα μας φαίνεται ελκυστικότερο από ένα χρηματικό κέρδος στο μέλλον, ακομα και αν το μελλοντικό κέρδος είναι μεγαλύτερο.

* Πολλές φορές προτιμούμε να αναβάλουμε να πληρώσουμε κάτι σήμερα, ακόμα και αν η αποπληρωμή του στο μέλλον θα μας κοστίσει στην πραγματικότητα περισσότερο.


Έστω τώρα ότι μελετούμε τη συμπεριφορά ενός ατόμου, του Φάνη. Ο Φάνης πρέπει να εκτελέσει μια σειρά εργασιών, ξεκινώντας από μια εργασία $s$ και καταλήγοντας ολοκληρώνοντας μια εργασία $t$. Για να φτάσει στην $t$ πρέπει να εκτελέσει μια σειρά από ενδιάμεσες εργασίες, με συγκεκριμένη σειρά μεταξύ τους. Επίσης, μπορεί να είναι δυνατόν να το κάνει αυτό με περισσότερες από μία σειρά εργασιών. Για παράδειγμα, δείτε τον παρακάτω γράφο (από το άρθρο των Kleinberg και Oren, βλ. στο τέλος), όπου οι κόμβοι είναι οι εργασίες, και τα βάρη στις ακμές δείχνουν το κόστος (π.χ. σε ημέρες) που χρειάζεται ο Φάνης για να προχωρήσει από μια εργασία στην επόμενη.

<img src="example_graph_1.png" width="400"/>

Ο Φάνης μπορεί να ξεκινήσει από το σημείο $s$ και να φτάσει στο σημείο $t$ με τους εξής τρόπους και κόστη:
* $s \rightarrow a \rightarrow b \rightarrow t$, με κόστος $16 + 2 + 2 = 20$.
* $s \rightarrow c \rightarrow d \rightarrow t$, με κόστος $8 + 8 + 8 = 24$.
* $s \rightarrow c \rightarrow e \rightarrow t$, με κόστος $8 + 2 + 16 = 26$.

Αν λοιπόν δράσει με βάση τη λογική, θα επιλέξει το μονοπάτι $s \rightarrow a \rightarrow b \rightarrow t$.

Στην πραγματικότητα όμως ο Φάνης πλανάται και μεροληπτεί υπέρ του παρόντος. Έτσι, όταν βρίσκεται σε έναν κόμβο, εκτιμά το κόστος των συνδέσμων που πρόσκεινται στον κόμβο με βάση το πραγματικό τους κόστος (γιατί συμβαίνει τώρα), αλλά εκτιμά το κόστος των κατοπινών συνδέσμων υποβαθμίζοντάς το κατά έναν παράγοντα $b < 1$, ώστε αν το πραγματικό κόστος είναι $c$, του Φάνη θα του φανεί ότι είναι $bc$.

Ας δούμε λοιπόν τι μονοπάτι θα ακολουθήσει ο Φάνης, αν $b = 1/2$.
1. Ξεκινάει από τον κόμβο $s$. Υπολογίζει το κόστος των μονοπατιών ως εξής:
  * $s \rightarrow a \rightarrow b \rightarrow t$, με κόστος $16 + (1/2)\times 2 + (1/2)\times 2 = 18$.
  * $s \rightarrow c \rightarrow d \rightarrow t$, με κόστος $8 + (1/2)\times 8 + (1/2)\times 8 = 16$.
  * $s \rightarrow c \rightarrow e \rightarrow t$, με κόστος $8 + (1/2)\times 2 + (1/2)\times 16 = 17$.
  
  Έτσι αποφασίζει να πάρει το *δεύτερο* μονοπάτι.
  
2. Τώρα βρίσκεται στον κόμβο $c$. Υπολογίζει εκ νέου το κόστος των μονοπατιών από το $c$ μέχρι το $t$ ως εξής:
  * $c \rightarrow d \rightarrow t$, με κόστος $8 + (1/2)\times 8 = 12$.
  * $c \rightarrow e \rightarrow t$, με κόστος $2 + (1/2)\times 16 = 10$.
  
  Έτσι αποφασίζει να αλλάξει, και να πάει στον κόμβο $e$, ακολουθώντας συνεπώς το *τρίτο* μονοπάτι.
  
3. Από τον κόμβο $e$ είναι μονόδρομος έως το $t$.

Βλέπουμε λοιπόν ότι ο Φάνης δεν θα επιλέξει το μονομάτι $s \rightarrow a \rightarrow b \rightarrow t$ με κόστος 20, αλλά το μονοπάτι $s \rightarrow c \rightarrow e \rightarrow t$ με κόστος 26!


Σκοπός της εργασίας είναι, αν σας δίνεται ένας γράφος όπως ο παραπάνω, και μία παράμετρος $b$, να υπολογίζετε:
* το βέλτιστο μονοπάτι
* το μονοπάτι που πραγματικά θα επιλέξει ο Φάνης

Ένας τρόπος να το βρούμε αυτό, είναι να ακολουθήσουμε τα παρακάτω στάδια:

1. Εύρεση όλων των μονοπατιών από το $s$ στο $t$.
2. Εύρεση του πραγματικού κόστους κάθε μονοπατιού από το $s$ στο $t$ (και εντοπισμός του βέλτιστου μονοπατιού).
3. Εκκίνηση από το $s$.
4. Μέχρι να φτάσουμε στο $t$:
  1. Υπολογισμός του κόστους κάθε μονοπατιού από τον κόμβο που βρισκόμαστε μέχρι το $t$, λαμβάνοντας υπόψη τη μεροληψία του παρόντος.
  2. Μεταφορά στον κόμβο που αντιστοιχεί στο καλύτερο μονοπάτι που βρήκαμε στο βήμα 4.A.
  3. Επανάληψη από το βήμα 4.

Για να μπορέσουμε να υλοποιήσουμε τα στάδια αυτά, πρέπει να ξέρουμε πώς να βρίσκουμε όλα τα μονοπάτια μεταξύ δύο κόμβων ενός γράφου. Αυτό είναι εύκολο να γίνει, με μια παραλλαγή της κατά βάθος αναζήτησης.

Για παράδειγμα, το σχετικό αποθετήριο του διδάσκοντα θα ονομαζόταν `louridas-algo-assignments` και θα ήταν προσβάσιμο στο https://github.com/dmst-algorithms-course/louridas-algo-assignments. * Τυχόν άλλα αποθετήρια απλώς θα αγνοηθούν*.

* Μέσα στο αποθετήριο αυτό θα πρέπει να δημιουργηθεί ένας κατάλογος `assignment-2018-1`.

* Μέσα στον παραπάνω κατάλογο το πρόγραμμα θα πρέπει να αποθηκευτεί με το όνομα `present_bias.py`.

* Δεν επιτρέπεται η χρήση έτοιμων βιβλιοθηκών γράφων ή τυχόν έτοιμων υλοποιήσεων των αλγορίθμων, ή τμημάτων αυτών. Επιτρέπεται η χρήση δομών δεδομένων της Python όπως στοίβες, λεξικά, κ.λπ.

* Το πρόγραμμα θα πρέπει να είναι γραμμένο σε Python 3.

* Το πρόγραμμα θα το καλεί ως χρήστης ως εξής (όπου `python` η κατάλληλη εντολή στο εκάστοτε σύστημα):

```
python present_bias.py <input_file> <bias_parameter> <start_node> <end_node>
```

Στην εντολή αυτή:
  * Η παράμετρος `<input_file>` είναι ένα όνομα αρχείου που περιέχει το γράφο που θα εξετάσουμε
  * Η παράμετρος `<bias_parameter>` είναι η παράμετρος $b$. 
  * Η παράμετρος `<start_node>` είναι ο κόμβος εκκίνησης.
  * Η παράμετρος `<end_node>` είναι ο τελικός κόμβος.

Τα αρχεία των γράφων θα έχουν την εξής μορφή:
```
s a 16
s c 8
a b 2
b t 2
c d 8
d t 8
c e 2
e t 16
```
Βλέπουμε δηλαδή ότι αποτελούνται από γραμμές, όπου η κάθε γραμμή δηλώνει τη σύνδεση δύο κόμβων και το βάρος που έχει ο σχετικός σύνδεσμος. 

Η έξοδος του προγράμματος θα είναι δύο γραμμές:
* Στην πρώτη γραμμή θα εμφανίζεται το βέλτιστο μονοπάτι και το κόστος του.
* Στη δεύτερη γραμμή θα εμφανίζεται το μονοπάτι που θα ακολουθήσει ο Φάνης και το κόστος του.

## Παραδείγματα Εκτέλεσης
 
### Παράδειγμα 1
 
Αν ο χρήστης καλέσει το πρόγραμμα για το αρχείο [example_graph_1.txt](example_graph_1.txt), με τον ακόλουθο τρόπο:

```
python present_bias.py example_graph_1.txt 0.5 s t
```

η έξοδος του προγράμματος θα πρέπει να είναι *ακριβώς* η παρακάτω:

```
['s', 'a', 'b', 't'] 20
['s', 'c', 'e', 't'] 26
```

### Παράδειγμα 2

Αν ο χρήστης καλέσει το πρόγραμμα για το αρχείο [example_graph_2.txt](example_graph_2.txt), με τον ακόλουθο τρόπο:

```
python present_bias.py example_graph_2.txt 0.5 start end
```
η έξοδος του προγράμματος θα πρέπει να είναι *ακριβώς* η παρακάτω:

```
['start', '5', '6', '7', '11', 'end'] 53
['start', '5', '6', '7', '11', '8', 'end'] 56
```

### Παράδειγμα 3

Ένα ενδιαφέρον φαινόμενο παρουσιάζεται στον παρακάτω γράφο:

<img src="example_graph_3.png" width="500"/>

Ο γράφος αυτός προσομοιάζει το τι γίνεται όταν αναβάλουμε κάτι, μέρα με την ημέρα. Έστω ότι οι κόμβοι από το 0 έως και το 6 είναι ημέρες και ο κόμβος $e$ αντιστοιχεί σε κάτι που πρέπει να κάνουμε, μία υποχρέωσή μας. Κάθε ημέρα έχουμε δύο επιλογές:

* Να εκπληρώσουμε την υποχρέωσή μας. Το κόστος είναι το ίδιο για κάθε ημέρα, 10.

* Να το αφήσουμε για την επόμενη. Η αναβολή έχει κόστος 1.

Είναι φανερό, ότι το καλύτερο που έχει να κάνει ο Φάνης, προκειμένου να πάει από το 0 στο e, είναι να διασχίσει τον απευθείας σύνδεσμο, δηλαδή να μην το αναβάλει καθόλου. Αλλά αν $b = 0{,}8$, τότε κάθε ημέρα θα το αφήνει για την επόμενη, έτσι ώστε θα εξαντλήσει την προθεσμία που έχει, με το μέγιστο κόστος.

Πράγματι, αν τρέξετε το πρόγραμμά σας για το αρχείο [example_graph_3.txt](example_graph_3.txt), που αντιστοιχεί στον παραπάνω γράφο, μπορείτε να εξερευνήσετε τα διάφορα σενάρια:

```
python present_bias.py example_graph_3.txt 0.5 0 e

['0', 'e'] 10
['0', '1', '2', '3', '4', '5', '6', 'e'] 16

python3 present_bias.py example_graph_3.txt 0.8 0 e
['0', 'e'] 10
['0', '1', '2', '3', '4', '5', '6', 'e'] 16

python3 present_bias.py example_graph_3.txt 0.9 0 e

['0', 'e'] 10
['0', 'e'] 10
```

## Για Περισσότερες Πληροφορίες

* Αν θέλετε μια ευκολοδιάβαστη εισαγωγή στις εφαρμογές των Συμπεριφορικών Οικονομικών, δείτε το βιβλίο: Richard H. Thaler and Cass R. Susstein, Nudge: Improving Decisions About Health, Wealth and Happiness, Penguin, 2009.

* Για μια εισαγωγή στις εφαρμογές των γράφων σε προβλήματα ασυνεπούς συμπεριφοράς δείτε το άρθρο: Jon Kleinberg and Sigal Oren, Time-inconsistent planning: a computational problem in behavioral economics, Communications of the ACM 61, 3 (February 2018), 99-107. DOI: https://doi.org/10.1145/3176189. Από το άρθρο αυτό προέρχεται και ο πρώτος γράφος παράδειγμα, και η έμπνευση για την παρούσα εργασία.

* Το θέμα της αναβλητικότητας στα οικονομικά το έχει πραγματευτεί ο Νομπελίστας George Akerlof, βλ.: George A. Akerlof, Procrastination and obedience, American Economic Review, 81, 2 (May 1991), 1-19.

* Για τους τρόπους με τους οποίους αλλάζουμε την εκτίμησή μας κατά μήκος του χρόνου, βλ.: Shane Frederick, George Loewenstein and Ted O'Donoghue, Time Discounting and Time Preference: A Critical Review, Journal of Economic Literature, 40, 2 (Junε 2002), pp. 351-401.
